```python
from writhe_tools.writhe import Writhe
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["text.usetex"] = True
plt.rcParams["font.family"] = "Times New Roman"
#plt.rcParams["mathtext.fontset"] = "custom"
#plt.rcParams["mathtext.rm"] = "Times New Roman"
```

## A2T & cluster 0

```python
fig, axes = plt.subplots(1,5, figsize=(15,3))
ax = axes.flat

for i in range(1,6):
    writhe = Writhe.load("a2t/cluster_0/a2t_writhe_data_dict_length_{}.pkl".format(i))
    writhe.plot_writhe_matrix(index=None, ax=ax[i-1], label_stride=8)  #xticks=residues, yticks=residues, to match example

fig.tight_layout()
plt.savefig('figures/A2T_cluster0.png',dpi=800, bbox_inches='tight')
```
