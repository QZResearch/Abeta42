```python
from writhe_tools.writhe import Writhe
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
plt.rcParams["text.usetex"] = True
plt.rcParams["font.family"] = "Times New Roman"
#plt.rcParams["mathtext.fontset"] = "custom"
#plt.rcParams["mathtext.rm"] = "Times New Roman"
```

```python
def getsimilarity(f1_1,f1_2,f2_1,f2_2):

    writhe_features1_1 = loadwrithe_features(f1_1)
    writhe_features1_2 = loadwrithe_features(f1_2)

    #print(writhe_features1_1.ravel().shape)
    #print(writhe_features1_2.ravel().shape)
    
    writhe_features1 = np.append(writhe_features1_1.ravel(), writhe_features1_2.ravel())
    #print(writhe_features1.shape)
    
    writhe_features2_1 = loadwrithe_features(f2_1)
    writhe_features2_2 = loadwrithe_features(f2_2)

    writhe_features2 = np.append(writhe_features2_1.ravel(), writhe_features2_2.ravel())

    similarity = matsimilarity(writhe_features1, writhe_features2)
    return similarity
```

```python
def loadwrithe_features(filename):
    writhe = Writhe.load(filename)
    mat = writhe.writhe_features
    mat_mean = np.mean(mat, axis=0)
    writhe_features = writhe.matrix(writhe_features=mat_mean)
    return writhe_features

def matsimilarity(mat1, mat2):
    assert (mat1.shape==mat2.shape), "shape of mat1 and mat2 is inconsistent"
    mat1_flatten = mat1.flatten()
    mat2_flatten = mat2.flatten()
    d = np.linalg.norm(mat1_flatten - mat2_flatten)
    return d#np.exp(-d**2)
```

```python
output_a2v_a2t = []
for i in range(4):
    f_a2v_1 = 'a2v/cluster_{}/a2v_writhe_data_dict_length_{}.pkl'.format(i,1)
    f_a2v_5 = 'a2v/cluster_{}/a2v_writhe_data_dict_length_{}.pkl'.format(i,5)
    for j in range(5):
        f_a2t_1 = 'a2t/cluster_{}/a2t_writhe_data_dict_length_{}.pkl'.format(j,1)
        f_a2t_5 = 'a2t/cluster_{}/a2t_writhe_data_dict_length_{}.pkl'.format(j,5)

        #writhe_wt = loadwrithe_features(f_wt)
        #writhe_a2v = loadwrithe_features(f_a2v)
        #f_wt = np.append([f_wt_1.ravel(),f_wt_5.ravel()])
        #f_a2v = np.append([f_a2v_1.ravel(),f_a2v_5.ravel()])
        #score = matsimilarity(writhe_a2v, writhe_wt)
        score = getsimilarity(f_a2v_1,f_a2v_5, f_a2t_1,f_a2t_5)
        #print('WT {} - A2V {}, score {:.4f}'.format(i, j, score))
        output_a2v_a2t.append([i,j,score])

        print('{:.3f} & '.format(score),end='')
    print('')
output_a2v_a2t = np.array(output_a2v_a2t)
```
